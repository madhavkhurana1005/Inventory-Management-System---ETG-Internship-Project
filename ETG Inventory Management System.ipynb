```python
import pandas as pd
import json

# Importing necessary libraries
```

```python
file = open('Inventory_data.json', 'r')
inv = file.read()
file.close()

Inventory = json.loads(inv)

# Retrieving data from json file - Inventory_data.json'
```

```python
Inventory

# Glimpse of Inventory data 
```

```python
final_amount = 0
Total_Sale = 0
Total_Bills = 0
wait_time = []                 # defining variables

def get_order():              # This is primary function which will take orders into the system 
    
    global order,qty,final_amount
    amount = 0 
    order = str(input("Enter Product ID : "))
    if (int(order) >= 125033 and int(order) <= 125063):                    # Taking only valid Product IDs
        qty = int(input("Item quantity : "))
        if (qty > Inventory[order]['Quantity Avail']):
                print("Only",int(Inventory[order]['Quantity Avail']), "Quantity Available")       # Inventory warning 
        else :
            display_items()
            Inventory[order]['Quantity Avail'] = Inventory[order]['Quantity Avail'] - qty
            amount = (int(Inventory[order]['price (Rs)']) * qty)
            final_amount += amount                                             # Calculations and manipulations
            wait_time.append(int(Inventory[order]['Time to prepare (min)']))
            ano = int(input("Want to add more Items? Enter 1 otherwise 0 : "))   # To add next product in the order
            print("\n")
            if (ano == 1):
                get_order()
            else:
                give_bill() 
                

    else :
        print("Invalid Product ID")
        get_order()
        
    
def display_items():                          # This function will display items ordered with various details as follows.
    items = 0
    print("************************************************")
    print("Item Name : ",Inventory[order]['Name'])
    print("Size : ",Inventory[order]['Size'])
    print("Rating : ",Inventory[order]['rating (out of 5)'],"/5 stars")
    print("Price : Rs",Inventory[order]['price (Rs)'], "per plate")
    print("Amount : Rs",Inventory[order]['price (Rs)']*qty)
    print("************************************************")
    items += 1 
        
def give_bill():
    global Total_Bills,Total_Sale                # This function will generate final Bill
    
    print("Your Bill Amount is : Rs", final_amount)
    print("You have to wait",max(wait_time)," minutes.")
    print("\n")
    print("Thank you for choosing us!")
    Total_Sale += final_amount 
    Total_Bills += 1
    
    print("----------------------------------------------------------------------------------------------------------")


def Inventory_left():                  # this function will show the total inventory present to the restaurant staff
    for i in Inventory:
        print(Inventory[i]['Quantity Avail'],'-',Inventory[i]['Name'],'-',Inventory[i]['Size'])
        
def Sales_data():                     # this function will show the total sales done by the restaurant
    for i in Inventory:
        print((10 - Inventory[i]['Quantity Avail'])*Inventory[i]['price (Rs)'],'-',Inventory[i]['Name'],'-',Inventory[i]['Size'])
    
def Show_menu():                     # this function will show the menu to staff for customer assistance
    for i in Inventory:
        print(Inventory[i]['price (Rs)'],'\t',Inventory[i]['rating (out of 5)'],'\t',Inventory[i]['Name'],'(',Inventory[i]['Size'],')')
    
def Download_Inventory_data():          # this function will download the inventory left to the staff system in json format
    a = []                              # This data can be further used for next day inventory planning and managemnt.
    for i in Inventory :
        a.append(Inventory[i]['Quantity Avail'])
        a1 = pd.Series(a)

    b = []
    for i in Inventory :
        b.append(Inventory[i]['Name'])
        b1 = pd.Series(b)

    c = []
    for i in Inventory :
        c.append(Inventory[i]['Size'])
        c1 = pd.Series(c)

    Net_invs = pd.concat([a1, b1, c1], axis=1)
    
    Inventory_left = Net_invs.to_json(orient='index')
    il = json.dumps(Inventory_left)
    il1 = open('Inventory_left.json','w')
    il1.write(il)
    il1.close()

    
def Download_Sales_data():                  #this function will download the sales data to the staff system in json format
    f = []                                  # This data can be further used for storage and analysis.
    for i in Inventory :
        f.append((10 - Inventory[i]['Quantity Avail'])*Inventory[i]['price (Rs)'])
        f1 = pd.Series(f)

    d = []
    for i in Inventory :
        d.append(Inventory[i]['Name'])
        d1 = pd.Series(d)

    e = []
    for i in Inventory :
        e.append(Inventory[i]['Size'])
        e1 = pd.Series(e)

    Net_Sale = pd.concat([f1, d1, e1], axis=1)
    
    Sales_data = Net_Sale.to_json(orient='index')
    sd = json.dumps(Sales_data)
    sd1 = open('sales_data.json','w')
    sd1.write(sd)
    sd1.close()
```
